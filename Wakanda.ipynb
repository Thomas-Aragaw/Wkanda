```python
import random
import ast
class Wakanda():
    def __init__(self):
        #opening the file that contains countries name, capital and currecny
        with open('countries.txt') as countries_object:
            self.data =countries_object.read()
        #reading a file from a dicitionay using ast.literal_eval() method
        self.countries_capitals = ast.literal_eval(self.data)
        self.result = 0
        self.high ='highest_score.txt'
        self.highest_score=0
    # creating a list of the key to index the countries
    def get_allcountries(self):
        allcountries = list(self.countries_capitals.keys())
        return allcountries
    # Retriving ranom country and then its list of values(capitals and countries)
    def get_the_items(self):
        country = random.choice(list(self.countries_capitals.keys()))
        the_list = self.countries_capitals[country]
        return country,the_list
    #Retriving the capital of the random country
    def get_random_country(self):
        country,the_list = self.get_the_items()
        capital= the_list[0]
        return country, capital
    #Retriving the currency of the random country
    def get_random_currency(self):
        country,the_list = self.get_the_items()
        currency= the_list[1]
        return country, currency
    # Retriving the curecny of a specific country
    def get_random_all(self,given):
        for key,value in self.countries_capitals.items():
            if key== given:
                the_currency = value[1]
                return the_currency
    #retriving the highest score from a file
    def get_highscore(self): 
        with open(self.high) as file_object:
            self.highest_score = file_object.read()
            self.highest_score = self.highest_score.strip()
            return self.highest_score
    #update the highest score
    def write_highscore(self,point): 
        with open(self.high, 'w')as file_object:
            file_object.write(point)
    #display the player name, available chance, Highest score and player score
    def display(self,Player_name,score,count):
        highest_score = self.get_highscore()
        highest_score = int(highest_score)
        start= f"{Player_name} your have {count} life \nThe highest score is {highest_score} \n{Player_name} your score is {score}"
        return start
    # check if user score is higher than the highest score
    def get_smartest(self,Player_name,score):
        highest_score = self.get_highscore()
        highest_score = int(highest_score)
        if score > highest_score:
            highest_score = self.write_highscore(str(score))
            smart = f"congratulations {Player_name} you are the smartest player"
            return smart
        end= f"Good bye {Player_name}"
        return end
```
